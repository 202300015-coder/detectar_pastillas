```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def ver(img, titulo="", cmap=None, tam=(6, 6)):
    plt.figure(figsize=tam)
    if len(img.shape) == 2:
        plt.imshow(img, cmap=cmap or "gray")
    else:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(titulo)
    plt.axis("off")
    plt.show()
```

```python
ruta = 'pills\green_capsule.jpg'
img = cv2.imread(ruta)

# x1, y1, x2, y2 -> AJUSTAR a tu área de trabajo real
x1, y1, x2, y2 = 190, 160, 405, 380

zona = img[y1:y2, x1:x2].copy()

ver(img, "Imagen completa", tam=(8, 8))
ver(zona, "Área de trabajo", tam=(5, 5))
```
