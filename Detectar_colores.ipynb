### Reto 1

```python
#rangos
Azul_B    = np.array([ 80, 100, 100], np.uint8)
Azul_A    = np.array([125, 255, 255], np.uint8)

Rojo1_B   = np.array([  0, 100, 100], np.uint8)
Rojo1_A   = np.array([  5, 255, 255], np.uint8)
Rojo2_B   = np.array([170, 100, 100], np.uint8)
Rojo2_A   = np.array([180, 255, 255], np.uint8)

Amarillo_B = np.array([ 20,  60,  80], np.uint8)
Amarillo_A = np.array([ 20, 255, 255], np.uint8)

Naranja_B  = np.array([  8, 100, 100], np.uint8)
Naranja_A  = np.array([ 20, 255, 255], np.uint8)

Verde_B    = np.array([ 40,  50,  50], np.uint8)
Verde_A    = np.array([ 80, 255, 255], np.uint8)

Morado_B   = np.array([105,  70,  40], np.uint8)
Morado_A   = np.array([135, 255, 170], np.uint8)

img3 = cv2.GaussianBlur(img_cont2.copy(), (5, 5), 0)
hsv3 = cv2.cvtColor(img3, cv2.COLOR_BGR2HSV)

mask_azul     = cv2.inRange(hsv3, Azul_B, Azul_A)
mask_rojo     = cv2.add(cv2.inRange(hsv3, Rojo1_B, Rojo1_A),
                        cv2.inRange(hsv3, Rojo2_B, Rojo2_A))
mask_amarillo = cv2.inRange(hsv3, Amarillo_B, Amarillo_A)
mask_naranja  = cv2.inRange(hsv3, Naranja_B, Naranja_A)
mask_verde    = cv2.inRange(hsv3, Verde_B, Verde_A)
mask_morado   = cv2.inRange(hsv3, Morado_B, Morado_A)

resultado3 = img3.copy()

for cont, _ in [cv2.findContours(mask_azul,     cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)]:
    cv2.drawContours(resultado3, cont, -1, (0, 0, 255), 3)

for cont, _ in [cv2.findContours(mask_rojo,     cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)]:
    cv2.drawContours(resultado3, cont, -1, (0, 255, 255), 3)

for cont, _ in [cv2.findContours(mask_amarillo, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)]:
    cv2.drawContours(resultado3, cont, -1, (255, 0, 0), 3)

for cont, _ in [cv2.findContours(mask_naranja,  cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)]:
    cv2.drawContours(resultado3, cont, -1, (0, 165, 255), 3)

for cont, _ in [cv2.findContours(mask_verde,    cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)]:
    cv2.drawContours(resultado3, cont, -1, (0, 255, 0), 3)

for cont, _ in [cv2.findContours(mask_morado,   cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)]:
    cv2.drawContours(resultado3, cont, -1, (255, 0, 255), 3)

resultado3_rgb = cv2.cvtColor(resultado3, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
plt.title('Contorno2.png  (original)')
plt.xticks([])
plt.yticks([])
plt.subplot(1, 2, 2)
plt.imshow(resultado3_rgb)
plt.title('Contornos: azul→rojo | rojo→amarillo | amarillo→azul')
plt.xticks([])
plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
img4 = cv2.GaussianBlur(img_ej2.copy(), (5, 5), 0)
hsv4 = cv2.cvtColor(img4, cv2.COLOR_BGR2HSV)

mask4_azul     = cv2.inRange(hsv4, Azul_B, Azul_A)
mask4_rojo     = cv2.add(cv2.inRange(hsv4, Rojo1_B, Rojo1_A),
                         cv2.inRange(hsv4, Rojo2_B, Rojo2_A))
mask4_amarillo = cv2.inRange(hsv4, Amarillo_B, Amarillo_A)

resultado4 = img4.copy()

for cont, _ in [cv2.findContours(mask4_azul,     cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)]:
    cv2.drawContours(resultado4, cont, -1, (255, 0, 0), 3)     # borde azul

for cont, _ in [cv2.findContours(mask4_rojo,     cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)]:
    cv2.drawContours(resultado4, cont, -1, (0, 0, 255), 3)     # borde rojo

for cont, _ in [cv2.findContours(mask4_amarillo, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)]:
    cv2.drawContours(resultado4, cont, -1, (0, 255, 255), 3)   # borde amarillo

resultado4_rgb = cv2.cvtColor(resultado4, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img4, cv2.COLOR_BGR2RGB))
plt.title('Ejemplo2.png  (original)')
plt.xticks([])
plt.yticks([])
plt.subplot(1, 2, 2)
plt.imshow(resultado4_rgb)
plt.title('Contornos por color detectado')
plt.xticks([])
plt.yticks([])
plt.tight_layout()
plt.show()
```

### Reto 2 — Contornos por color

```python
def cont_col(img, tit, bk=(5,5), am=120, ck=(5,5)):
    if bk == (1,1):
        sv = img.copy()
    else:
        sv = cv2.GaussianBlur(img, bk, 0)

    hsv = cv2.cvtColor(sv, cv2.COLOR_BGR2HSV)

    msk = {
        'a': (cv2.inRange(hsv, Azul_B, Azul_A), (255, 200, 0)),
        'r': (cv2.add(cv2.inRange(hsv, Rojo1_B, Rojo1_A),
                      cv2.inRange(hsv, Rojo2_B, Rojo2_A)), (0, 255, 0)),
        'y': (cv2.inRange(hsv, Amarillo_B, Amarillo_A), (128, 0, 255)),
        'n': (cv2.inRange(hsv, Naranja_B, Naranja_A), (0, 165, 255)),
        'v': (cv2.inRange(hsv, Verde_B, Verde_A), (0, 200, 100)),
        'm': (cv2.inRange(hsv, Morado_B, Morado_A), (255, 0, 255)),
    }

    res = img.copy()
    k1 = np.ones((3,3), np.uint8)

    for _, (m, cb) in msk.items():
        m = cv2.morphologyEx(m, cv2.MORPH_OPEN, k1)

        if ck != (1,1):
            m = cv2.morphologyEx(m, cv2.MORPH_CLOSE, np.ones(ck, np.uint8))

        cnt, _ = cv2.findContours(m, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        cnt_u = [c for c in cnt if cv2.contourArea(c) > am]
        cv2.drawContours(res, cnt_u, -1, cb, 4)

    return cv2.cvtColor(res, cv2.COLOR_BGR2RGB), tit

r1, t1 = cont_col(img_fig, 'Figuras.png', bk=(1,1), am=150, ck=(1,1))
r2, t2 = cont_col(img_suav, 'Suavizado.png', bk=(7,7), am=80, ck=(5,5))

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
for ax, o, t, r in [
    (axs[0,0], cv2.cvtColor(img_fig, cv2.COLOR_BGR2RGB), 'Figuras.png (original)', None),
    (axs[0,1], None, t1, r1),
    (axs[1,0], cv2.cvtColor(img_suav, cv2.COLOR_BGR2RGB), 'Suavizado.png (original)', None),
    (axs[1,1], None, t2, r2),
]:
    if r is None:
        ax.imshow(o)
    else:
        ax.imshow(r)

    ax.set_title(t, fontsize=9)
    ax.set_xticks([])
    ax.set_yticks([])

plt.suptitle('Reto 2 - Contornos por color  |  azul=cyan  rojo=verde  amarillo=magenta  naranja=naranja  verde=verde  morado=magenta', fontsize=8)
plt.tight_layout()
plt.show()
```

```python
colores_camara = {
    'Azul': ((Azul_B, Azul_A), (255, 0, 0)),
    'Rojo': (((Rojo1_B, Rojo1_A), (Rojo2_B, Rojo2_A)), (0, 0, 255)),
    'Amarillo': ((Amarillo_B, Amarillo_A), (0, 255, 255)),
    'Verde': ((Verde_B, Verde_A), (0, 255, 0)),
    'Morado': ((Morado_B, Morado_A), (255, 0, 255)),
}

cam = cv2.VideoCapture(1)


kernel = np.ones((5, 5), np.uint8)

while True:
    ret, frame = cam.read()

    frame = cv2.flip(frame, 1)
    suavizado = cv2.GaussianBlur(frame, (5, 5), 0)
    hsv = cv2.cvtColor(suavizado, cv2.COLOR_BGR2HSV)
    salida = frame.copy()

    for nombre, (rangos, color_bgr) in colores_camara.items():
        if nombre == 'Rojo':
            mask = cv2.add(
                cv2.inRange(hsv, rangos[0][0], rangos[0][1]),
                cv2.inRange(hsv, rangos[1][0], rangos[1][1])
            )
        else:
            mask = cv2.inRange(hsv, rangos[0], rangos[1])

        mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
        mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
        contornos, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        for contorno in contornos:
            area = cv2.contourArea(contorno)
            if area < 500:
                continue

            x, y, w, h = cv2.boundingRect(contorno)
            cv2.rectangle(salida, (x, y), (x + w, y + h), color_bgr, 2)
            cv2.putText(salida, nombre, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color_bgr, 2, cv2.LINE_AA)

    cv2.imshow('Colores', salida)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cam.release()
cv2.destroyAllWindows()
```
